### 初步使用方法

```python
import igraph as ig
```

```python
# 以指定节点数量和连边列表的方式创建一个简单图
g = ig.Graph(n=4, edges=[[0, 1], [0, 2], [1, 3], [2, 3]])
print(g)
ig.plot(g, bbox=(300, 300))
```

```python
# 创建一个空图，并为其增加节点和连边
g = ig.Graph()
# 增加节点和连边
g.add_vertices(3) # 增加三个节点0，1，2
g.add_edges([(0,1), (0,2), (1,2)])
print(g)
ig.plot(g, bbox=(300, 300))
```

```python
# 使用字典指定边权值(或任何其他顶点/边属性)
g = ig.Graph(
    n=4, edges=[[0, 1], [2, 3]],
    edge_attrs={'weight': [0.1, 0.2]},
    vertex_attrs={'color': ['b', 'g', 'g', 'y']}
)
```

```python
# 生成库中自带的经典网络（在第02课中会详细讲到）
g = ig.Graph.Famous('Zachary')
ig.plot(g, bbox=(300, 300))
# # 以及其他经典图
# g = ig.Graph.Full(n=3)
# g = ig.Graph.Lattice(dim=[3, 3], circular=False)
# g = ig.Graph.Ring(n=4, circular=False)
# g = ig.Graph.GRG(100, 0.2)
```
