## igraph编程实践基础教程第02课：使用库函数生成经典的网络模型

```python
# !/usr/bin/python3
# -*- coding: utf-8 -*-
# Author ： 单哥的科研日常（公众号、B站）
# 示例程序所用库的版本：igraph==0.10.4
```

```python
import igraph as ig
print(ig.__version__)
```

```python
# 生成完全网络
g = ig.Graph.Full(n=10)
ig.plot(g, bbox=(300, 300))
```

```python
g = ig.Graph.Full(n=10, directed=True) # 有向
ig.plot(g, bbox=(300, 300))
```

```python
# 生成一维环
g = ig.Graph.Ring(n=10, circular=True) # circular表示是否使用周期性边界条件
ig.plot(g, bbox=(300, 300))
```

```python
from IPython.display import Image
# K近邻耦合网络
Image(filename = './attached files/K_Regular.png', width=600, height=300)
```

```python
g = ig.Graph.K_Regular(n=10, k=4)
ig.plot(g, layout="circle", bbox=(300, 300))
```

```python
# 星型网络
Image(filename = './attached files/Star.png', width=600, height=300)
```

```python
g = ig.Graph.Star(10)
ig.plot(g, bbox=(300, 300))
```

```python
# ER随机网络
Image(filename = './attached files/ER.png', width=600, height=300)
```

```python
# g = ig.Graph.Erdos_Renyi(n=20, p=0.3)  # 类似于NetworkX中的gnp模型
g = ig.Graph.Erdos_Renyi(n=20, m=30)  # 类似于NetworkX中的gnm模型
# 直接写g = ig.Graph.Erdos_Renyi(20, 60)会报错：InternalError: Error at src/games/erdos_renyi.c:53: Invalid probability given. -- Invalid value
ig.plot(g, bbox=(300, 300))
```

```python
# WS小世界网络
Image(filename = './attached files/WS.png', width=600, height=300)
```

```python
g = ig.Graph.Watts_Strogatz(dim=1,size=20, nei=2, p=0.1)
ig.plot(g, layout="circle", bbox=(300, 300))
```

```python
# BA无标度网络
Image(filename = './attached files/BA.png', width=600, height=300)
```

```python
g = ig.Graph.Barabasi(n=50,m=2)
ig.plot(g, bbox=(300, 300))
```

```python
# 随机增长网络
Image(filename = './attached files/Growing_Random.png', width=600, height=300)
```

```python
n = 100
m = 3
g = ig.Graph.Growing_Random(n=n, m=m)
ig.plot(g, bbox=(300, 300))
```

```python
# 随机块网络
Image(filename = './attached files/SBM.png', width=600, height=300)
```

```python
n = 60
block_sizes = [20, 15, 25]
pref_matrix = [[0.25, 0.01, 0.01], [0.01, 0.35, 0.02], [0.01, 0.02, 0.40]]
g = ig.Graph.SBM(n, pref_matrix, block_sizes)
# 设置标签
g.vs["label"] = range(len(g.vs))
ig.plot(g, bbox=(300, 300))
```

```python
# 随机几何网络
Image(filename = './attached files/GRG.png', width=600, height=300)
```

```python
g = ig.Graph.GRG(100, 0.2)
ig.plot(g, vertex_size=10, bbox=(300, 300))
```

```python
# 生成具有指定幂律度分布的非增长无标度网络
Image(filename = './attached files/Static_Power_Law.png', width=600, height=300)
```

```python
g = ig.Graph.Static_Power_Law(100, 200, exponent_out=2.5, exponent_in=-1) # exponent_in为负值，说明为无向网络
ig.plot(g, vertex_size=10, bbox=(300, 300))
```

```python
g = ig.Graph.Static_Power_Law(100, 200, exponent_out=2.5, exponent_in=2.1) # 有向
ig.plot(g, vertex_size=10, bbox=(300, 300))
```

```python
# 生成具有给定度序列的图
Image(filename = './attached files/Degree_Sequence.png', width=600, height=300)
```

```python
g = ig.Graph.Degree_Sequence(out=[2,3,1,5,8,3,3,4,5,6])
ig.plot(g, vertex_size=10, bbox=(300, 300))
```
