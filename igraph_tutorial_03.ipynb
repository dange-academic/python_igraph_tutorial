```python
import igraph as ig
print(ig.__version__)
```

#### 从邻接矩阵创建一个图

```python
# 无权情况
g = ig.Graph.Adjacency([[0, 1, 1], [1, 0, 1], [1, 1, 0]])
```

```python
# 加权情况
wg = ig.Graph.Weighted_Adjacency([[0, 2, 1], [2, 0, 0], [1, 0, 0]])
```

#### 将图保存为文件，如邻接矩阵或连边

```python
from IPython.display import Image
Image(filename = './attached files/write_adjacency.png', width=600, height=400)
```

```python
g = ig.Graph.Erdos_Renyi(n=20, p=0.3, directed=False)
g.write_adjacency("./attached files/AD.dat", eol='\n')
# 或者csv的格式
g.write_adjacency("./attached files/AD.csv", sep=',')
```

```python
Image(filename = './attached files/write_edgelist.png', width=600, height=400)
```

```python
# 保存为连边的格式
g.write_edgelist("./attached files/edgelist.dat")
```

#### 加载外部文件生成图

```python
# 加载邻接矩阵
Image(filename = './attached files/Read_Adjacency.png', width=600, height=400)
```

```python
g = ig.Graph.Read_Adjacency("./attached files/AD.dat", directed=False)
```

```python
# 加载连边
Image(filename = './attached files/Read_Edgelist.png', width=600, height=400)
```

```python
g = ig.Graph.Read_Edgelist("./attached files/edgelist.dat", directed=False)   # 注意：igraph读取网络连边数据时要确保节点标签从0开始顺序编号
ig.plot(g, bbox=(300, 300))
```

```python
# 更多文件的读取方式见官网：
Image(filename = './attached files/igraph_read_write.png', width=600, height=400)
```

```python
# 读取Gephi的图文件，如gml格式
g = ig.Graph.Read_GML("./attached files/network.gml")
ig.plot(g, vertex_size=10, bbox=(300, 300))
```

```python
# 将其他库生成的图转化为igraph下的图：igraph可以读写networkx和graph-tool图格式
import networkx as nx
print(nx.__version__)
```

```python
nwx = nx.gnm_random_graph(20, 40)
g = ig.Graph.from_networkx(nwx)
```

```python
nx.draw(nwx)
```

```python
ig.plot(g, vertex_size=10, bbox=(300, 300))
```
